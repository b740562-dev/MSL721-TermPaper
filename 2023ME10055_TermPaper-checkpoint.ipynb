```python
# staying near to the college helps to get four year of extra education at very minimal cost 
# =============================================================================
# STEP 3 — NAIVE OLS (Baseline Model)
# =============================================================================
print("\n" + "=" * 65)
print(" STEP 3: NAIVE OLS — BASELINE MODEL")
print("=" * 65)

X_cols = ['educ', 'exper', 'expersq', 'black', 'south', 'urban', 'married']
y = df['lwage']
X = sm.add_constant(df[X_cols])

ols_naive = sm.OLS(y, X).fit()

print("\nNaive OLS Results:")
print(f"{'Variable':<15} {'Coef':>10} {'Std Err':>10} {'t-stat':>10} {'p-value':>10}")
print("-" * 58)
for var in ols_naive.params.index:
    p = ols_naive.pvalues[var]
    sig = '***' if p < 0.01 else '**' if p < 0.05 else '*' if p < 0.1 else ''
    print(f"{var:<15} {ols_naive.params[var]:>10.4f} {ols_naive.bse[var]:>10.4f}"
          f" {ols_naive.tvalues[var]:>10.3f} {p:>10.4f} {sig}")
print(f"\nR-squared  : {ols_naive.rsquared:.4f}")
print(f"Adj R²     : {ols_naive.rsquared_adj:.4f}")
print(f"F-statistic: {ols_naive.fvalue:.2f}  (p = {ols_naive.f_pvalue:.6f})")
print(f"\n→ Naive OLS: 1 extra year of education raises wages by {ols_naive.params['educ']*100:.2f}%")
print("  BUT this estimate is likely BIASED — we'll prove it below.")
```

```python
# =============================================================================
# STEP 4 — PROBLEM 1: MULTICOLLINEARITY
# =============================================================================
print("\n" + "=" * 65)
print(" STEP 4: DETECTING & FIXING MULTICOLLINEARITY")
print("=" * 65)

print("""
THEORY:
Multicollinearity occurs when independent variables are highly correlated.
In this dataset: age ≈ educ + exper + 6 (by construction!).
This inflates standard errors → makes coefficients unreliable.

Detection tool: Variance Inflation Factor (VIF)
  VIF = 1/(1 - R²_j) where R²_j = R² from regressing X_j on all other X's
  Rule of thumb:  VIF > 5  → concern
                  VIF > 10 → serious multicollinearity
""")

# Test with age INCLUDED (problematic model)
X_with_age = sm.add_constant(df[X_cols + ['age']])
vif_data_bad = pd.DataFrame({
    'Variable': X_with_age.columns,
    'VIF': [variance_inflation_factor(X_with_age.values, i)
            for i in range(X_with_age.shape[1])]
}).query("Variable != 'const'").sort_values('VIF', ascending=False)

print("VIF — WITH age included (problematic):")
print(f"{'Variable':<15} {'VIF':>10}  {'Status'}")
print("-" * 42)
for _, row in vif_data_bad.iterrows():
    status = "🚨 SEVERE" if row['VIF'] > 10 else "⚠  WARNING" if row['VIF'] > 5 else "✓  OK"
    print(f"{row['Variable']:<15} {row['VIF']:>10.2f}  {status}")
```

```python
# Fix: drop age (it's a linear combo of educ + exper)
X_clean = sm.add_constant(df[X_cols])
vif_data_good = pd.DataFrame({
    'Variable': X_clean.columns,
    'VIF': [variance_inflation_factor(X_clean.values, i)
            for i in range(X_clean.shape[1])]
}).query("Variable != 'const'").sort_values('VIF', ascending=False)

print("\nVIF — WITHOUT age (age dropped — fix applied):")
print(f"{'Variable':<15} {'VIF':>10}  {'Status'}")
print("-" * 42)
for _, row in vif_data_good.iterrows():
    status = "🚨 SEVERE" if row['VIF'] > 10 else "⚠  WARNING" if row['VIF'] > 5 else "✓  OK"
    print(f"{row['Variable']:<15} {row['VIF']:>10.2f}  {status}")
```

```python
# Correlation between educ and age
r_educ_age = df['educ'].corr(df['age'])
r_educ_exp = df['educ'].corr(df['exper'])
print(f"\nCorrelation between educ & age  : {r_educ_age:.3f}")
print(f"Correlation between educ & exper: {r_educ_exp:.3f}")
print("\n→ FIX APPLIED: Dropped 'age' from the model.")
print("  age is near-perfectly determined by educ + exper (identity relation).")
print("  Note: exper & expersq have high VIF by design (x² is always correlated with x).")
print("  This is expected for polynomial terms — we retain both for economic validity.")
print("  All non-polynomial VIF values are below 5 — multicollinearity resolved.")

# VIF bar chart
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("Multicollinearity Diagnosis — Variance Inflation Factor (VIF)",
             fontsize=14, fontweight='bold')

colors_bad  = [RED if v > 10 else AMBER if v > 5 else GREEN for v in vif_data_bad['VIF']]
colors_good = [RED if v > 10 else AMBER if v > 5 else GREEN for v in vif_data_good['VIF']]

axes[0].barh(vif_data_bad['Variable'], vif_data_bad['VIF'], color=colors_bad, edgecolor='white')
axes[0].axvline(5, color=AMBER, linestyle='--', linewidth=1.5, label='VIF = 5 (warning)')
axes[0].axvline(10, color=RED,  linestyle='--', linewidth=1.5, label='VIF = 10 (severe)')
axes[0].set_title('Before Fix (age included)')
axes[0].set_xlabel('VIF')
axes[0].legend(fontsize=8)

axes[1].barh(vif_data_good['Variable'], vif_data_good['VIF'], color=colors_good, edgecolor='white')
axes[1].axvline(5, color=AMBER, linestyle='--', linewidth=1.5, label='VIF = 5 (warning)')
axes[1].axvline(10, color=RED,  linestyle='--', linewidth=1.5, label='VIF = 10 (severe)')
axes[1].set_title('After Fix (age dropped)')
axes[1].set_xlabel('VIF')
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.savefig('sem7', dpi=140, bbox_inches='tight')
plt.close()
print("\n[Plot saved] 02_multicollinearity.png")
```

```python
# =============================================================================
# STEP 5 — PROBLEM 2: HETEROSKEDASTICITY
# =============================================================================
print("\n" + "=" * 65)
print(" STEP 5: DETECTING & FIXING HETEROSKEDASTICITY")
print("=" * 65)

print("""
THEORY:
Heteroskedasticity = the variance of the error term is NOT constant.
It violates OLS Assumption 5 (homoskedasticity).
Consequence: OLS coefficients are still unbiased, BUT standard errors
are WRONG → invalid t-tests and confidence intervals.

Detection tests:
  1. Breusch-Pagan test : regresses squared residuals on regressors
  2. White's test       : more general, also checks interactions
  H0 (null): errors are homoskedastic
  Reject H0 if p-value < 0.05 → heteroskedasticity confirmed.
""")

residuals = ols_naive.resid
fitted    = ols_naive.fittedvalues
```

```python
# Test 1: Breusch-Pagan
bp_stat, bp_pval, _, _ = het_breuschpagan(residuals, X)
print(f"Breusch-Pagan Test:")
print(f"  LM statistic : {bp_stat:.4f}")
print(f"  p-value      : {bp_pval:.6f}  {'→ REJECT H0: Heteroskedasticity detected!' if bp_pval < 0.05 else '→ Fail to reject H0'}")

# Test 2: White's test
white_stat, white_pval, _, _ = het_white(residuals, X)
print(f"\nWhite's Test:")
print(f"  LM statistic : {white_stat:.4f}")
print(f"  p-value      : {white_pval:.6f}  {'→ REJECT H0: Heteroskedasticity detected!' if white_pval < 0.05 else '→ Fail to reject H0'}")
```

```python
# error terms are linearly related with x as confirmed by both breusch pagan and white test.


# Visual check: residuals vs fitted
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
fig.suptitle("Heteroskedasticity Diagnosis & Fix", fontsize=14, fontweight='bold')

# (a) Residuals vs Fitted
axes[0,0].scatter(fitted, residuals, alpha=0.3, s=12, color=BLUE)
axes[0,0].axhline(0, color=RED, linewidth=1.5, linestyle='--')
axes[0,0].set_xlabel('Fitted values')
axes[0,0].set_ylabel('Residuals')
axes[0,0].set_title('(a) Residuals vs Fitted\n(Fanning pattern = heteroskedasticity)')
z = np.polyfit(fitted, residuals, 2)
p_fit = np.poly1d(z)
x_line = np.linspace(fitted.min(), fitted.max(), 200)
axes[0,0].plot(x_line, p_fit(x_line), color=RED, linewidth=1.5)

# (b) Scale-Location plot (sqrt |residuals| vs fitted)
sqrt_abs_resid = np.sqrt(np.abs(residuals))
axes[0,1].scatter(fitted, sqrt_abs_resid, alpha=0.3, s=12, color=GREEN)
axes[0,1].set_xlabel('Fitted values')
axes[0,1].set_ylabel('√|Residuals|')
axes[0,1].set_title('(b) Scale-Location Plot\n(Non-flat trend = heteroskedasticity)')
z2 = np.polyfit(fitted, sqrt_abs_resid, 1)
axes[0,1].plot(x_line, np.poly1d(z2)(x_line), color=RED, linewidth=2)

# (c) Q-Q plot of residuals
sm.qqplot(residuals, line='s', ax=axes[1,0], alpha=0.5, markersize=3)
axes[1,0].set_title('(c) Q-Q Plot of Residuals\n(Deviations from line = non-normality)')
axes[1,0].get_lines()[1].set_color(RED)

# (d) Residuals vs Education (key variable)
axes[1,1].scatter(df['educ'], residuals, alpha=0.3, s=12, color=AMBER)
axes[1,1].axhline(0, color=RED, linewidth=1.5, linestyle='--')
axes[1,1].set_xlabel('Education (years)')
axes[1,1].set_ylabel('Residuals')
axes[1,1].set_title('(d) Residuals vs Education\n(Wider spread at high education = problem)')
educ_groups = df.groupby('educ').apply(lambda x: residuals[x.index].std())
axes[1,1].plot(educ_groups.index, educ_groups.values * 1.5, color=RED, linewidth=2, label='±1.5 SD')
axes[1,1].plot(educ_groups.index, -educ_groups.values * 1.5, color=RED, linewidth=2)
axes[1,1].legend(fontsize=8)

plt.tight_layout()
plt.savefig('sem8', dpi=140, bbox_inches='tight')
plt.close()
print("\n[Plot saved] 03_heteroskedasticity.png")
```

```python
# THE FIX: HC3 Robust Standard Errors (White's heteroskedasticity-consistent SE)---we want our statistical inferences to account for reality of heteroskedasticity
print("\nFIX: Applying HC3 Robust Standard Errors (White, 1980)")
print("     HC3 formula: e_i² / (1 - h_ii)²  where h_ii = leverage")
print("     This corrects SE without assuming any particular form of heteroskedasticity.\n")

ols_robust = sm.OLS(y, X).fit(cov_type='HC3')

print(f"{'Variable':<15} {'OLS Coef':>10} {'Naive SE':>10} {'Robust SE':>10} {'Change %':>10}")
print("-" * 58)
for var in ols_naive.params.index:
    naive_se  = ols_naive.bse[var]
    robust_se = ols_robust.bse[var]
    change    = (robust_se - naive_se) / naive_se * 100
    flag = " ← SE changed!" if abs(change) > 10 else ""
    print(f"{var:<15} {ols_naive.params[var]:>10.4f} {naive_se:>10.4f} "
          f"{robust_se:>10.4f} {change:>+9.1f}%{flag}")

print("\n→ FIX APPLIED: HC3 robust standard errors are now used for all inference.")
print("  Coefficients unchanged — only the standard errors (and thus t-stats/p-values) differ.")
```

```python
# =============================================================================
# STEP 6 — PROBLEM 3: ENDOGENEITY
# =============================================================================
print("\n" + "=" * 65)
print(" STEP 6: DETECTING & FIXING ENDOGENEITY")
print("=" * 65)

print("""
THEORY:
Endogeneity occurs when a regressor is correlated with the error term.
Here: educ is endogenous because of ABILITY BIAS.

  True model: lwage = β₀ + β₁·educ + β₂·ability + ε
  Problem   : ability is unobserved → goes into ε
  Since ability ↑ → both educ ↑ and wage ↑, Cov(educ, ε) > 0
  Effect    : OLS β̂₁ is BIASED DOWNWARD (underestimates true return)

FIX: Instrumental Variables (IV) / 2SLS
  We need an instrument Z such that:
    1. RELEVANCE  : Cov(Z, educ) ≠ 0   (Z affects education)
    2. EXOGENEITY : Cov(Z, ε) = 0      (Z doesn't affect wages directly)

  Our instrument: nearc4 (grew up near a 4-year college)
    Relevance : Living near a college lowers cost of attending → more education ✓
    Exogeneity: College proximity shouldn't affect wages directly (only via education) ✓
                (Card 1995 argues family location is quasi-random)
""")
```

```python
# Hausman Test for Endogeneity
print("HAUSMAN TEST for endogeneity:")
print("  H0: educ is exogenous (OLS is consistent)")
print("  H1: educ is endogenous (IV is needed)\n")

# Stage 1: regress educ on instrument + controls
stage1_vars = ['nearc4', 'exper', 'expersq', 'black', 'south', 'urban', 'married']
X_stage1 = sm.add_constant(df[stage1_vars])
stage1 = sm.OLS(df['educ'], X_stage1).fit()

print(f"First Stage (educ ~ nearc4 + controls):")
print(f"  Coefficient on nearc4 : {stage1.params['nearc4']:.4f}  (t = {stage1.tvalues['nearc4']:.3f})")
print(f"  F-statistic (1st stage): {stage1.fvalue:.2f}  [Rule of thumb: F > 10 for strong instrument]")
print(f"  → Instrument is {'STRONG ✓' if stage1.fvalue > 10 else 'WEAK ✗'}")

# Hausman: add first-stage residuals to original OLS, test significance
v_hat = stage1.resid  # first-stage residuals
X_hausman = sm.add_constant(pd.concat([df[X_cols], v_hat.rename('v_hat')], axis=1))
hausman_reg = sm.OLS(y, X_hausman).fit(cov_type='HC3')
hausman_p   = hausman_reg.pvalues['v_hat']
hausman_t   = hausman_reg.tvalues['v_hat']

print(f"\n  Coefficient on v_hat : {hausman_reg.params['v_hat']:.4f}")
print(f"  t-statistic          : {hausman_t:.3f}")
print(f"  p-value              : {hausman_p:.4f}")
print(f"  → {'REJECT H0: Endogeneity confirmed → use IV/2SLS' if hausman_p < 0.1 else 'Fail to reject H0: OLS may be OK'}")

# 2SLS using linearmodels
print("\n2SLS ESTIMATION (Two-Stage Least Squares):")
endog  = df[['educ']]
exog   = sm.add_constant(df[['exper', 'expersq', 'black', 'south', 'urban', 'married']])
instru = df[['nearc4']]

iv_model = IV2SLS(dependent=df['lwage'],
                  exog=exog,
                  endog=endog,
                  instruments=instru).fit(cov_type='robust')

print(f"\n{'Variable':<15} {'Coef':>10} {'Std Err':>10} {'t-stat':>10} {'p-value':>10}")
print("-" * 58)
for var in iv_model.params.index:
    p = iv_model.pvalues[var]
    sig = '***' if p < 0.01 else '**' if p < 0.05 else '*' if p < 0.1 else ''
    print(f"{var:<15} {iv_model.params[var]:>10.4f} {iv_model.std_errors[var]:>10.4f}"
          f" {iv_model.tstats[var]:>10.3f} {p:>10.4f} {sig}")
```

```python
# =============================================================================
# STEP 7 — FINAL COMPARISON TABLE
# =============================================================================
print("\n" + "=" * 65)
print(" STEP 7: RESULTS COMPARISON — ALL MODELS")
print("=" * 65)

educ_ols_naive  = ols_naive.params['educ']
educ_ols_robust = ols_robust.params['educ']
educ_iv         = iv_model.params['educ']

se_ols_naive  = ols_naive.bse['educ']
se_ols_robust = ols_robust.bse['educ']
se_iv         = iv_model.std_errors['educ']

print(f"""
╔══════════════════════════════════════════════════════════╗
║          RETURN TO EDUCATION — COEFFICIENT ON educ       ║
╠══════════════════════════════════════════════════════════╣
║  Model                     Coef (%)    Std Error        ║
║ ─────────────────────────────────────────────────────── ║
║  (1) Naive OLS              {educ_ols_naive*100:>6.2f}%     ({se_ols_naive:.4f})       ║
║  (2) OLS + Robust SE        {educ_ols_robust*100:>6.2f}%     ({se_ols_robust:.4f})       ║
║  (3) IV/2SLS (nearc4)       {educ_iv*100:>6.2f}%     ({se_iv:.4f})       ║
╠══════════════════════════════════════════════════════════╣
║  Bias correction: {(educ_iv - educ_ols_naive)*100:+.2f}% (2SLS vs Naive OLS)          ║
╚══════════════════════════════════════════════════════════╝
""")

print("INTERPRETATION:")
print(f"  Naive OLS says : 1 more year of education → wages up by {educ_ols_naive*100:.2f}%")
print(f"  2SLS (true)    : 1 more year of education → wages up by {educ_iv*100:.2f}%")
print(f"  The {(educ_iv - educ_ols_naive)*100:.2f}% gap is ABILITY BIAS — OLS underestimates")
print(f"  the true return because ability (omitted) is correlated with education.")

# Coefficient comparison chart
fig, ax = plt.subplots(figsize=(10, 5))
fig.suptitle("Return to Education — Coefficient Comparison Across Models",
             fontsize=13, fontweight='bold')

models = ['(1) Naive OLS', '(2) OLS +\nRobust SE', '(3) IV/2SLS\n(endogeneity fixed)']
coefs  = [educ_ols_naive, educ_ols_robust, educ_iv]
ses    = [se_ols_naive, se_ols_robust, se_iv]
colors = [RED, AMBER, GREEN]

bars = ax.bar(models, [c * 100 for c in coefs], color=colors,
              alpha=0.85, width=0.5, edgecolor='white')
ax.errorbar(models, [c * 100 for c in coefs],
            yerr=[s * 196 for s in ses],
            fmt='none', color='black', capsize=8, linewidth=2, capthick=2)

for bar, c in zip(bars, coefs):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{c*100:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=11)

ax.set_ylabel('% increase in wage per extra year of education')
ax.set_title('Error bars = 95% confidence intervals\n'
             'Green bar (IV/2SLS) is the unbiased estimate', fontsize=10)
ax.set_ylim(0, max(c * 100 for c in coefs) + 4)

ax.annotate('OLS underestimates\ndue to ability bias',
            xy=(0, educ_ols_naive * 100), xytext=(0.6, educ_ols_naive * 100 + 3),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

plt.tight_layout()
plt.savefig('sem9', dpi=140, bbox_inches='tight')
plt.close()
print("\n[Plot saved] 04_coefficient_comparison.png")
```
